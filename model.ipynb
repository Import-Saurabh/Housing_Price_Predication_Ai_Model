```python
import pandas as pd
import numpy as np

data=pd.read_csv('Bengaluru_House_Data.csv')
data.head()
data.shape
data.info()
data.isna().sum()
```

```python
data.drop(columns=['area_type','availability','society','balcony'],inplace=True)
```

```python
data.describe()
```

```python
data['location'].value_counts()
```

```python
data['location']=data['location'].fillna('Sarjapur Road')
```

```python
data['size'].value_counts()
```

```python
data['size']=data['size'].fillna('2 BHK')
```

```python
data['bath']=data['size'].fillna(data['bath'].median())
```

```python
data['bhk'] = data['size'].str.split().str.get(0).fillna(0).astype(int)
data[data.bhk>20]
```

```python
data['total_sqft'].unique()
```

```python
def convertRange(x):
 temp=x.split('-')
 if len(temp)==2:
        return(float(temp[0]) + float(temp[1]))/2
 try :
        return float(x)
 except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convertRange)
```

```python
data.head()
```

```python
data['price_per_sqft'] = data['price'] * 100000 / data['total_sqft']
data['price_per_sqft']
```

```python
data.describe()
```

```python
data['location'].value_counts()
```

```python
data['location']=data['location'].apply(lambda x: x.strip())
location_count=data['location'].value_counts()
```

```python
location_count_less_10 = location_count[location_count<=10]
location_count_less_10
```

```python
data['location']=data['location'].apply(lambda x:'other' if x in location_count_less_10 else x)
```

```python
data['location'].value_counts()
```

```python
data=data[((data['total_sqft']/data['bhk'])>=300)]
data.shape
```

```python
 data.price_per_sqft.describe()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_output = pd.concat([df_output, gen_df], ignore_index=True)

    return df_output

data = remove_outliers_sqft(data)
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices=np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats={}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)

    return df.drop(exclude_indices,axis='index')
```

```python
data=bhk_outlier_remover(data)
```

```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
data
```

```python
columns_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),
    remainder='passthrough'
)
scaler=StandardScaler()
lr=LinearRegression()
# Create a pipeline with column transformer, standard scaler, and linear regression
pipe = make_pipeline(columns_trans,scaler,lr)

# Fit the entire pipeline on the training data
pipe.fit(x_train, y_train)

# Predict using the fitted pipeline
y_pred_lr = pipe.predict(x_test)

# Print the R-squared score
print("R-squared score:", r2_score(y_test, y_pred_lr))
```

```python
from sklearn.linear_model import Lasso

# Initialize Lasso regression
lasso = Lasso(alpha=1.0)  # You can adjust alpha as needed

# Fit Lasso using the same pipeline pattern
pipe_lasso = make_pipeline(columns_trans, scaler, lasso)
pipe_lasso.fit(x_train, y_train)

# Predict
y_pred_lasso = pipe_lasso.predict(x_test)

# Print R-squared score
from sklearn.metrics import r2_score
print("R-squared score (Lasso):", r2_score(y_test, y_pred_lasso))
```

```python
from xgboost import XGBRegressor
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

# Initialize XGBoost regressor
xgb = XGBRegressor(random_state=0, n_estimators=100, learning_rate=0.1)

# Create pipeline
pipe_xgb = make_pipeline(columns_trans, scaler, xgb)

# Fit on training data
pipe_xgb.fit(x_train, y_train)

# Predict
y_pred_xgb = pipe_xgb.predict(x_test)

# Print R-squared score
print("R-squared score (XGBoost):", r2_score(y_test, y_pred_xgb))
```
